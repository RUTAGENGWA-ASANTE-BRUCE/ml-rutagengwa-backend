```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
IMG_SIZE=70
from keras.preprocessing import image
import cv2
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
DATADIR="C:/Users/pc/Desktop/coding/MachineLearning/Bruce_ML_Project/backend/datasets"
pickle_in=open(f"{DATADIR}/bruce.pickle","rb")
initial_data=pickle.load(pickle_in)
def get_training_data():
    all_features=[]
    all_labels=[]

    for features, labels in initial_data:
        all_labels.append(labels)
        all_features.append(features)
    
#     X=all_features/255.0
    
    X=np.array(all_features)
    Y=np.array(all_labels)
    return {"X":X,"Y":all_labels}
data=get_training_data()

print(data["X"].shape)
print(len(data['Y']))
print(len(data["X"]))
print(np.array(data["X"][0]).shape)
data["X"].shape[1:]
print(np.expand_dims(data["X"][0],0).shape)
```

```python
import cv2
import tensorflow as tf
import os
def resizeImage(myFrame):
    new_array=cv2.resize(myFrame,(IMG_SIZE,IMG_SIZE))
    return new_array
class_names=[]
for dataset in os.listdir("C:/Users/pc/Desktop/coding/MachineLearning/Bruce_ML_Project/backend/datasets"):
    class_names.append(dataset[0:-7])
model=tf.keras.models.load_model(os.path.join("C:/Users/pc/Desktop/coding/MachineLearning/Bruce_ML_Project/backend","BRUCE_ML_MODEL.model"))
images_folder="C:/Users/pc/Desktop/coding/MachineLearning/Bruce_ML_Project/backend/TEST_IMAGES_FOLDER"
for image_name in os.listdir(images_folder)[30:70]:
    img=cv2.imread(os.path.join(images_folder,image_name),cv2.COLOR_BGR2RGB)
    myImage= resizeImage(img)
    
    x=np.expand_dims(myImage/255.0,0)
    prediction=model.predict([x])
    print(prediction)
    predicted_class = class_names[np.argmax(prediction)]
    print(f"predicted class: {predicted_class}")
    plt.imshow(myImage,cmap="gray")
    plt.show()
    
```
